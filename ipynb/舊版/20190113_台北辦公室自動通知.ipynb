# Time

```python
import time
a = "Sat Mar 28 22:24:24 2016"
print(a)

unix = time.mktime(time.strptime(a,"%a %b %d %H:%M:%S %Y"))
print(unix)

# 格式化成Sat Mar 28 22:24:24 2016形式
print(time.strftime("%a %b %d %H:%M:%S %Y", time.localtime(unix)))
```

# 值日生編排

```python
import time
import pandas as pd

start='2018-08-13 08:00:00'
end='2021-08-17 08:00:00'

datestart=time.mktime(time.strptime(start,"%Y-%m-%d %H:%M:%S"))
dateend=time.mktime(time.strptime(end,"%Y-%m-%d %H:%M:%S"))

name = ['DAVE', '鄧秀明', '雅雅', '曹志成', 'jerrydeng', 'JSWANG']

dates = []
n = 0
while datestart<dateend:
    # 格式化成Sat Mar 28 22:24:24 2016形式
    date = time.strftime("%Y-%m-%d %H:%M:%S", time.localtime(datestart))
    #print(date)
    dates.append([date, False, "{0} 這週值日生是 {1} 詳細請參閱https://goo.gl/15qpxK 感謝 {1}!💖".format(date, name[n%6])])
    datestart += 86400*7
    n += 1
#print(dates)
df = pd.DataFrame(dates, columns=['date', 'bool', 'text'])
df['date'] = pd.to_datetime(df['date'])
#print(df.dtypes)

df
```

```python
# 存檔
df.to_csv('LINE Notify.txt', index=False, encoding='utf8')
```
